## Basic Plotting

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(sin(a))
```

```python
x = linspace(0, 2*pi, 50) # create linearly spaced array (start, end, #elements)
plt.plot(sin(x))
```

```python
plt.plot(x, sin(x), x, sin(2*x)) # ploting pairs of (x1,y1) (x2,y2)
```

```python
plt.plot(sin(x), 'bo') # blue circles. check out documentation.
```

```python
plt.plot(x, sin(x), 'b-o', x, sin(2*x), 'r-^')
```

```python
y = sin(x)
plt.scatter(x, y)
```

```python
x = random.rand(200) # from numpy.random
y = random.rand(200)
size = random.rand(200)*30
color = random.rand(200)
plt.scatter(x, y, size, color)
plt.colorbar()
```

## Multiple Plots

```python
plt.plot(a)
plt.plot(b)
```

```python
plt.figure()
plt.plot(a)
plt.figure()
plt.plot(b)
```

```python
x = array([1,2,3,2,1])
y = array([1,3,2,3,1])
plt.subplot(2, 1, 1) # rows, columns, active plot
# rows and columns are dimension of the plot
plt.plot(x)
plt.subplot(2, 1, 2)
plt.plot(y)
```

```python
f, axes = plt.subplots(2, 2, sharex=True, sharey=True)
axes # is an array of references to each subplot
```

```python
axes[0,0].plot(sin(linspace(0,2*pi,20)))
axes[0,1].plot(y)
axes[1,1].plot(2*sin(linspace(0,2*pi,20)))
axes[1,0].cla() # this clears the data from the axes
f
```

```python
plt.figure()
x = linspace(0, 2*pi, 50)
plt.plot(sin(x), label='sin')
plt.plot(cos(x), label='cos')
plt.legend(loc='best') # best location for legend
plt.xlabel('Radians')
plt.ylabel('Amplitude', fontsize='large')
plt.title('Sine and Cosine Curves')
plt.grid()
```

```python
plt.clf() # clears current figure but keeps the data
# plt.close() # close the current figure
# plt.close('all') # close all figures
```

## Image Display

```python
from scipy.misc import face
img = face()
plt.imshow(img, extent=[-25,25,-25,25])
```

```python
img = face(True)
plt.imshow(img, extent=[-25,25,-25,25])
plt.colorbar()
```

```python
plt.imshow(img, extent=[-25,25,-25,25], cmap=plt.cm.bone) # sets color mapping
plt.colorbar()
```
